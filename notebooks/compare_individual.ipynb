```python
first_dir = './data/label/1/'
second_dir = './data/label/2/'
out_dir = './data/stat/individual/'

infiles_1 = [first_dir + s + '.json' for s in filenames]
infiles_2 = [second_dir + s + '_LABEL.json' for s in filenames]
```

```python
for i in range(0, len(names)):
    print('Processing {0}'.format(names[i]))
    label_data_1 = []
    label_data_2 = []
    for j in range(0, len(filenames)):
        name, _ = filenames[j].split('-')
        if name == names[i]:
            label_data_1_tmp = json.load(open(infiles_1[j]))
            label_data_2_tmp = json.load(open(infiles_2[j]))
            if len(label_data_1_tmp) == len(label_data_2_tmp):
                label_data_1 = label_data_1 + label_data_1_tmp
                label_data_2 = label_data_2 + label_data_2_tmp

    if len(label_data_1) != len(label_data_2):
        print("{0}: Length not equal".format(filenames[i]))
        continue

    # Convert to numpy array
    np_label_data_1 = np.array(label_data_1)
    np_label_data_2 = np.array(label_data_2)

    # Extract starting and ending points
    start_1 = np.array([d['start'] for d in np_label_data_1[:,1]])
    start_2 = np.array([d['start'] for d in np_label_data_2[:,1]])
    end_1 = np.array([d['end'] for d in np_label_data_1[:,1]])
    end_2 = np.array([d['end'] for d in np_label_data_2[:,1]])

    # Calculate width
    width_1 = end_1 - start_1
    width_2 = end_2 - start_2

    # Calculate diff
    start_diff = np.abs(start_1 - start_2)
    end_diff = np.abs(end_1 - end_2)

    stat = {}

    stat['1'] = {}
    stat['1']['mean'] = np.mean(width_1)
    stat['1']['min'] = np.min(width_1)
    stat['1']['max'] = np.max(width_1)
    stat['1']['std'] = np.std(width_1)

    stat['2'] = {}
    stat['2']['mean'] = np.mean(width_2)
    stat['2']['min'] = np.min(width_2)
    stat['2']['max'] = np.max(width_2)
    stat['2']['std'] = np.std(width_2)

    stat['start_diff'] = {}
    stat['start_diff']['mean'] = np.mean(start_diff)
    stat['start_diff']['min'] = np.min(start_diff)
    stat['start_diff']['max'] = np.max(start_diff)
    stat['start_diff']['std'] = np.std(start_diff)

    stat['end_diff'] = {}
    stat['end_diff']['mean'] = np.mean(end_diff)
    stat['end_diff']['min'] = np.min(end_diff)
    stat['end_diff']['max'] = np.max(end_diff)
    stat['end_diff']['std'] = np.std(end_diff)

    fout = open(out_dir + names[i] + '_STAT.txt', 'w')
    fout.write("Label 1\n")
    fout.write("Mean: {0}\n".format(stat['1']['mean']))
    fout.write("Std: {0}\n".format(stat['1']['std']))
    fout.write("Min: {0}\n".format(stat['1']['min']))
    fout.write("Max: {0}\n".format(stat['1']['max']))
    fout.write("\n\n")

    fout.write("Label 2\n")
    fout.write("Mean: {0}\n".format(stat['2']['mean']))
    fout.write("Std: {0}\n".format(stat['2']['std']))
    fout.write("Min: {0}\n".format(stat['2']['min']))
    fout.write("Max: {0}\n".format(stat['2']['max']))
    fout.write("\n\n")

    fout.write("Starting Point Difference\n")
    fout.write("Mean: {0}\n".format(stat['start_diff']['mean']))
    fout.write("Std: {0}\n".format(stat['start_diff']['std']))
    fout.write("Min: {0}\n".format(stat['start_diff']['min']))
    fout.write("Max: {0}\n".format(stat['start_diff']['max']))
    fout.write("\n\n")

    fout.write("Ending Point Difference\n")
    fout.write("Mean: {0}\n".format(stat['end_diff']['mean']))
    fout.write("Std: {0}\n".format(stat['end_diff']['std']))
    fout.write("Min: {0}\n".format(stat['end_diff']['min']))
    fout.write("Max: {0}\n".format(stat['end_diff']['max']))

    fout.close()
print('Done.')
```
